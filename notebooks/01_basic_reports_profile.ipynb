```python
import pandas as pd
```

```python
BUFFER_DAYS = 90
path = "C:\\Users\\camac\\source\\SDPipe\\notebooks\\data\\raw\\pd_collisions_datasd.csv"
```

```python
#read data in
basic_collison_df = pd.read_csv(path, dtype={"report_id":"string"},parse_dates=['date_time'])
```

```python
basic_collison_df
```

```python
max_dt = basic_collison_df['date_time'].max()
cutoff = max_dt - pd.Timedelta(days=BUFFER_DAYS)
coll_90days = basic_collison_df[basic_collison_df['date_time'] >= cutoff].copy()
coll_90days
```

```python
#Grain validation:
coll_90days['report_id'].is_unique, coll_90days['report_id'].nunique(), len(coll_90days)
```

```python
nulls = (coll_90days.isna().mean().sort_values(ascending=False) * 100).round(2)
nulls
```
