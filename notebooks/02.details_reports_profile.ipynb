```python
import pandas as pd
BUFFER_DAYS = 90
path_basic_report = "C:\\Users\\camac\\source\\SDPipe\\notebooks\\data\\raw\\pd_collisions_datasd.csv"
path_details_report = "C:\\Users\\camac\\source\\SDPipe\\notebooks\\data\\raw\\pd_collisions_details_datasd.csv"
```

```python
details_df = pd.read_csv(path_details_report, dtype={"report_id":"string"},parse_dates=['date_time'])
basic_df = pd.read_csv(path_basic_report, dtype={"report_id":"string"},parse_dates=['date_time'])
```

```python
max_df = basic_df['date_time'].max()
cutoff = max_df - pd.Timedelta(days=BUFFER_DAYS)
basic_90 = basic_df[basic_df["date_time"] >= cutoff].copy()
details_90 = details_df[details_df['report_id'].isin(basic_90['report_id'])].copy()
basic_90.shape, details_90.shape
```

```python
rows_per_collision = details_90.groupby("report_id").size()
rows_per_collision.describe()
```
